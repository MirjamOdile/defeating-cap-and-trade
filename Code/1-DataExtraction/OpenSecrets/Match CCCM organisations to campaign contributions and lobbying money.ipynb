```python
import csv
indivs_names = []
with open('Analysis/indivs_names.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile)
    for row in spamreader:
        indivs_names.append(row[0])
        
indivs_Orgname = []
with open('Analysis/indivsOrgname.csv', newline='', encoding='windows-1252') as csvfile:
    spamreader = csv.reader(csvfile)
    for row in spamreader:
        indivs_Orgname.append(row[0])
indivs_Orgname = [i.lower() for i in indivs_Orgname]

indivs_UltOrg = []
with open('Analysis/indivsUltOrg.csv', newline='', encoding='windows-1252') as csvfile:
    spamreader = csv.reader(csvfile)
    for row in spamreader:
        indivs_UltOrg.append(row[0])
indivs_UltOrg = [i.lower() for i in indivs_UltOrg]
```

```python
# Extract direct matches
cccm_Orgname = [i for i in organisations if i in indivs_Orgname]
cccm_UltOrg = [i for i in organisations if i in indivs_UltOrg]
# Drop perfect matches from lists
indivs_Orgname = [i for i in indivs_Orgname if not i in cccm_Orgname]
indivs_UltOrg = [i for i in indivs_UltOrg if not i in cccm_UltOrg]
```

```python
# Extract fuzzy matches
for i, organisation in enumerate(organisations):
    switch = 0
    for j, name in enumerate(indivs_Orgname):
        ratio = fuzz.token_sort_ratio(organisation, name)
        if ratio > 80:
            if switch == 0:
                    print('----\n', organisation,'\n')
                    switch = 1
            else:
                continue
            print(i, ratio, ': ', name, '\n')
```
