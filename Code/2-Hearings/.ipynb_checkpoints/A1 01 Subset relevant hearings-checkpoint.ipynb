## Investigate the hearings that don't meet the selection criteria

```python
count_title = 0
for i, text in enumerate(t.texts):
    if len(re.findall('climate.?change|global.?warming', text['title'])) > 0:
        count_title += 1
print('Of the {} hearings, {} mention climate change or global warming in their title. These hearings will definitely be included in the data.\n\n'
      'Index\tcc\tc\tcp\ttitle'.format(len(t.texts), count_title))
        
for i, text in enumerate(t.texts):
    if len(re.findall('climate.?change|global.?warming', text['title'])) > 0:
        print(i, '\t', 
              len(text['keywords_climatechange']), '\t',
              len(text['keywords_climate']), '\t',
              len(text['keywords_carbonpricing']), '\t',
              text['title'])
```

```python
count = 0
for i, text in enumerate(t.texts):
    if (len(re.findall('climate.?change|global.?warming', text['title'])) == 0 and
        #len(text['keywords_climatechange']) > 1 and
        len(text['keywords_carbonpricing']) < 5):
        count += 1
print('Of the remaining {} hearings, {} mention carbon pricing keywords less than 5 times '
      'and will be removed.\n'
      'These are the hearings that will be removed:\n\n'
      'Index\tcc\tc\tcp\ttitle'.format(len(t.texts)-count_title, count))
        
for i, text in enumerate(t.texts):
    if (len(re.findall('climate.?change|global.?warming', text['title'])) == 0 and
        #len(text['keywords_climatechange']) > 1 and
        len(text['keywords_carbonpricing']) < 5 ):
        print(i, '\t', 
              len(text['keywords_climatechange']), '\t',
              len(text['keywords_climate']), '\t',
              len(text['keywords_carbonpricing']), '\t',
              text['title'])
```

```python
count = 0
for i, text in enumerate(t.texts):
    if (len(re.findall('climate.?change|global.?warming', text['title'])) != 0 or
        len(text['keywords_carbonpricing']) >= 5):
        count += 1
print('{} hearings mention a climate change keyword, as well as a climate change keyword in the title or 5+ carbon pricing keywords. \n'
      'These are the hearings that will constitute the data set for the following analysis:\n\n'
      'Index\tcc\tc\tcp\ttitle'.format(count))
        
for i, text in enumerate(t.texts):
    if (len(re.findall('climate.?change|global.?warming', text['title'])) != 0 or
        len(text['keywords_carbonpricing']) >= 5):
        print(i, '\t', 
              len(text['keywords_climatechange']), '\t',
              len(text['keywords_climate']), '\t',
              len(text['keywords_carbonpricing']), '\t',
              text['title'])
```

<br>
    
## Limit the hearings to the ones meeting the selection criteria

```python
t.texts = [text for text in t.texts if 
           len(re.findall('climate.?change|global.?warming', text['title'])) != 0 or
           len(text['keywords_carbonpricing']) >= 5]
        
len(t.texts)
```

<br>
    
## Check for appropriations / nomination hearings and annual reports

```python
count = 0
for i, text in enumerate(t.texts):
    if len(re.findall('appropriation', text['title'])) != 0:
        count += 1
print('{} hearings are appropriations hearings and will be removed:\n\n'
      'Index\ttitle'.format(count))
        
for i, text in enumerate(t.texts):
    if len(re.findall('appropriation', text['title'])) != 0:
        print(i, '\t', 
              text['title'])
```

```python
count = 0
for i, text in enumerate(t.texts):
    if len(re.findall('nomination', text['title'])) != 0:
        count += 1
print('{} hearings are nomination hearings and will be removed:\n\n'
      'Index\ttitle'.format(count))
        
for i, text in enumerate(t.texts):
    if len(re.findall('nomination', text['title'])) != 0:
        print(i, '\t',
              text['title'])
```

```python
count = 0
for i, text in enumerate(t.texts):
    if len(re.findall('annual report', text['title'])) != 0:
        count += 1
print('{} hearings is an annual report and will be removed:\n\n'
      'Index\ttitle'.format(count))
        
for i, text in enumerate(t.texts):
    if len(re.findall('annual report', text['title'])) != 0:
        print(i, '\t',
              text['title'])
```

<br>
    
## Remove appropriations and nomination hearings

```python
t.texts = [text for text in t.texts if 
           len(re.findall('appropriation|nomination|annual report', text['title'])) == 0]
        
len(t.texts)
```
